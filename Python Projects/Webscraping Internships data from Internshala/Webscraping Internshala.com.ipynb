```python
#Entering the url which is to be scraaped.
internshala_url="https://internshala.com/internships/content%20writing-internship"
#Creating a list with metrics we want to scrap.
list=['Digital Marketing','Tekolutions']
```

```python
scraper=AutoScraper()
result=scraper.build(internshala_url,list)
print(result)
```

```python
scraper.get_result_similar(internshala_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_l5v8':'Job_Title','rule_ks00':'Company'})
scraper.keep_rules(['rule_l5v8','rule_ks00'])
scraper.save('job_search')
```

```python
#Data is stored in dictionary named webscraped.
webscraped= scraper.get_result_similar('https://internshala.com/internships/graphic%20design-internship', group_by_alias=True)
webscraped['Company']
```

```python
# Converted dictionary into a dataframe to export it as a excel file. Scrapped graphic design related internships.
df= pd.DataFrame(webscraped)
df.to_csv('Internships_on_Internshala.csv')
df
```

```python
## Scrapping content writing related internships.
webscrap2= scraper.get_result_similar('https://internshala.com/internships/content%20writing-internship', group_by_alias=True)
df2=pd.DataFrame(webscrap2)
df2.to_csv('Content_writing_jobs.csv')
df2
```

```python
# Scrapping marketing related internships.
webscrap3= scraper.get_result_similar('https://internshala.com/internships/marketing-internship', group_by_alias=True)
df3=pd.DataFrame(webscrap3)
df3.to_csv('marketing_jobs.csv')
df3
```
